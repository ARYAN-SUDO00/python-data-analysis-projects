# Sales Data Analysis

```python
import pandas as pd
sales = pd.read_csv("train.csv")
sales.head(10)
```

```python
# Sales by category
category = round((sales.groupby("Category")["Sales"].sum()), 2)
category
```

```python
# Top 5 best sellers
sales["Sub-Category"].value_counts().head()
```

```python
# Indentify the month with the highest number of orders
sales["DateTime"] = pd.to_datetime(sales["Order Date"], dayfirst=True)
sales["Month"] = sales["DateTime"].dt.month_name()

monthly_orders = sales["Month"].value_counts()
highest_orders_month = monthly_orders.idxmax()
highest_orders = int(monthly_orders.max())
print(f"The month with the highest orders is {highest_orders_month} with the total orders of {highest_orders}")
```

```python
# Indentify the month with the highest sales
monthly_sales = sales.groupby("Month")["Sales"].sum()
highest_sales_month = monthly_sales.idxmax()
highest_sales = round(monthly_sales.max(), 2)
print(f"The month with the highest sales is {highest_sales_month} with the total sales of {highest_sales}$")
```

```python
# Calculate the total revenue
revenue = sales["Sales"].sum()
print(f"Total revenue is {round(revenue, 2)}$")
```
