# Netflix Movies & Shows Analysis

```python
import pandas as pd

netflix = pd.read_csv("netflix_titles.csv")
netflix["date_added"] = netflix["date_added"].str.strip()
netflix["date_added"] = pd.to_datetime(netflix["date_added"])
```

```python
# Shows the number of movies and tv shows
types = netflix["type"].value_counts()
print(types)
```

```python
# Lists top 5 most common genres
genres = netflix["listed_in"].str.split(", ").explode()
print(genres.value_counts().head())
```

```python
# Drop rows where date_added is null 
netflix_dates = netflix.dropna(subset=['date_added'])

# Get oldest and newest
old_idx = netflix_dates["date_added"].idxmin()
new_idx = netflix_dates["date_added"].idxmax()

oldest_content = netflix_dates.loc[old_idx, "title"]
newest_content = netflix_dates.loc[new_idx, "title"]

print(f"'{oldest_content}' was the first title added (on {netflix_dates.loc[old_idx, 'date_added'].date()})")
print(f"'{newest_content}' was the most recent title added (on {netflix_dates.loc[new_idx, 'date_added'].date()})")
```

```python
# Shows added per year
shows_per_year = netflix_dates["date_added"].dt.year.value_counts()
shows_per_year = shows_per_year.sort_index()
print(shows_per_year)
```
