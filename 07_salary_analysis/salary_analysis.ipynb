# Employees Salary Analysis

```python
import pandas as pd
salary = pd.read_csv("employee_salary.csv")
salary.head()
```

```python
# Average salary by department
salary_by_department = round((salary.groupby("Department")["Yearly_Salary(USD)"].mean()), 2)
salary_by_department
```

```python
highest = salary["Yearly_Salary(USD)"].idxmax()
print(f"The highest paying job is {salary.loc[highest, "Job_Title"]} from the {salary.loc[highest, "Department"]} department") 
```

```python
lowest = salary["Yearly_Salary(USD)"].idxmin()
print(f"The lowest paying job is {salary.loc[lowest, "Job_Title"]} from the {salary.loc[lowest, "Department"]} department") 
```

```python
# Average salary by gender
genders = round((salary.groupby("Gender")["Yearly_Salary(USD)"].mean()), 2)
genders
```

```python
total_salary = salary["Yearly_Salary(USD)"].sum()
print(f"The company spends {total_salary}$ per year on salaries")
```
