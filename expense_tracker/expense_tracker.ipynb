# Expense Tracker


```python
import pandas as pd
expenses = pd.read_csv("expenses.csv")
expenses.head(10)
```

```python
total_expenses = expenses["Amount(USD)"].sum()
print(f"Total spending of 2025 is {total_expenses}$")
```

```python
spending_by_category = expenses.groupby("Category")["Amount(USD)"].sum()
print(spending_by_category)
```

```python
highest_spending_day = expenses["Amount(USD)"].idxmax()
highest_amount = expenses.loc[highest_spending_day, "Amount(USD)"]
date = expenses["Date"][highest_spending_day]

print(f"The highest expense day of 2025 is {date} with an expense of {highest_amount}")
```

```python
expenses["Date"] = pd.to_datetime(expenses["Date"])
expenses["Month"] = expenses["Date"].dt.month_name()

months_order = ["January", "February", "March", "April", "May", "June", 
                "July", "August", "September", "October", "November", "December"]

expenses["Month"] = pd.Categorical(expenses["Month"], categories=months_order, ordered=True)
months = expenses.groupby("Month", observed=True)["Amount(USD)"].sum()
print(months)
```
