# Student Marks Analysis

```python
import pandas as pd
marks = pd.read_csv("students.csv")
marks.head()
```

```python
# Calculate the average marks of each student
marks["Average"] = marks.mean(axis=1, numeric_only=True)
highest_idx = marks["Average"].idxmax()
min_idx = marks["Average"].idxmin()
```

```python
# Find out the topper and the student who scored the lowest
top_student = marks.loc[highest_idx, "Name"]
last_student = marks.loc[min_idx, "Name"]

print(f"{top_student} scored the highest")
print(f"{last_student} scored the lowest")
```

```python
# Create a new file
marks.to_csv("new_students.csv", index=False)
```
