# Mobile Phone Sales Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
phones_data = pd.read_csv("Mobile phone price.csv")
```

```python
# Cleaning the dataset

phones_data.columns = phones_data.columns.str.strip()

phones_data["Price ($)"] = phones_data["Price ($)"].astype(str).str.replace(r'[\$,]', '', regex=True)

phones_data["RAM"] = phones_data["RAM"].str.extract(r'(\d+)').astype(int)
phones_data["Price ($)"] = pd.to_numeric(phones_data["Price ($)"], errors="coerce")

```

```python
# Strip Plot showing the relationship between RAM and mobile phone prices
sns.set_style("ticks")

sns.stripplot(data=phones_data, x="RAM", y="Price ($)",  hue="Brand", jitter=0.25, alpha=0.6, palette="tab20")
sns.despine()



plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Relationship between RAM and Mobile Phone Prices")
plt.xlabel('RAM(GB)')

plt.ylabel("Price(USD)")
plt.tight_layout()
plt.show()
```

```python
# Histogram showing the distribution of mobile phone prices
sns.set_style("white")
sns.histplot(data=phones_data, x="Price ($)", bins=15, color="steelblue", edgecolor="white")
sns.despine()

plt.title("Distribution of Mobile Phone Prices")
```
