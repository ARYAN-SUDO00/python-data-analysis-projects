## Word Frequency Analyzer

```python
import pandas as pd

# Open the text file
with open("article.txt", "r") as f:
    raw_text = f.read()
```

```python
# Clean the article
clean_text = raw_text.lower().replace(".", "").replace(",", "").replace('"', "").replace(":", "").replace(";", "")
words = clean_text.split()
```

```python
# Filter out the words with less than 4 letters
new_words = []
for word in words:
    if len(word) >= 4:
        new_words.append(word)

df_words = pd.DataFrame(new_words, columns=["word"])
```

```python
# Count the frequency of words and display 10 most occured words
frequency_count = df_words["word"].value_counts()
frequency_count.head(10)
```

```python
word = frequency_count.idxmax()
print(f"'{word}' is the most occuring word")
```
